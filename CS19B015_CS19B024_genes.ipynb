```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("assignmentcounts.csv")
```

```python
df.head()
```

MEAN vs Variance Graph

```python
gene_variance = df.var(axis = 1)
gene_mean = df.mean(axis = 1)
```

```python
plt.plot(gene_mean,gene_variance,'ro')
plt.xlabel('Mean')
plt.ylabel('Variance')
plt.yscale('log')
plt.xscale('log')
plt.title('Gene Count Mean vs Varinace')
plt.grid(True)
#plt.show()
plt.savefig('Mean_vs_Variance.png')
```
